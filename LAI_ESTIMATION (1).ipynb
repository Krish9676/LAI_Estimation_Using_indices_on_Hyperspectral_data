```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
data = pd.read_csv("/content/indices_LAI.csv")
x = data.drop(["Crop_stage", "observed_LAI"], axis=1)
y = data["observed_LAI"]
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Standardize the features (important for KNN)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Create and train the KNN model
k = 4  # Number of neighbors
knn_model = KNeighborsRegressor(n_neighbors=k)
knn_model.fit(x_train_scaled, y_train)
```

```python
# Make predictions on the test set
y_pred = knn_model.predict(x_test_scaled)
```

```python
# Evaluate the model
r2 = r2_score(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
```

```python
print(f"R^2: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red')
plt.xlabel('Actual LAI')
plt.ylabel('Predicted LAI')
plt.title(f'Best-Fit Model: KNN\nR-squared: {r2:.4f}, RMSE: {rmse:.4f}, MAE: {mae:.4f}')
plt.show()
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
# Load the data
data = pd.read_csv("/content/indices_LAI.csv")
x = data.drop(["Crop_stage", "observed_LAI"], axis=1)
y = data["observed_LAI"]
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Create and train the XGBoost model
xgb_model = XGBRegressor(n_estimators=35, learning_rate=0.1)  # You can adjust hyperparameters here
xgb_model.fit(x_train, y_train)
```

```python
# Make predictions on the test set
y_pred = xgb_model.predict(x_test)
```

```python
# Evaluate the model
r2 = r2_score(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
```

```python
print(f"R^2: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red')
plt.xlabel('Actual LAI')
plt.ylabel('Predicted LAI')
plt.title(f'Best-Fit Model: XGBOOST\nR-squared: {r2:.4f}, RMSE: {rmse:.4f}, MAE: {mae:.4f}')
plt.show()
```

```python
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    xgb_model = XGBRegressor(n_estimators=i, learning_rate=0.1)
    xgb_model.fit(x_train, y_train)
    y_pred = xgb_model.predict(x_test)
    r2 = r2_score(y_test, y_pred)
    print(f"Model R^2 score on test set: {r2:.4f}")
    print("")
```

```python
# Get feature importance scores from the trained model
feature_importance = xgb_model.feature_importances_

```

```python
# Plot feature importance for XGBoost
plt.figure(figsize=(8, 6))
plt.bar(x_train.columns, feature_importance)
plt.title("XGBoost Feature Importance")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
import lightgbm as lgb
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
# Load the data
data = pd.read_csv("/content/indices_LAI.csv")
x = data.drop(["Crop_stage", "observed_LAI"], axis=1)
y = data["observed_LAI"]
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Create and train the LightGBM model
lgb_model = lgb.LGBMRegressor(n_estimators=100, learning_rate=0.1)  # You can adjust hyperparameters here
lgb_model.fit(x_train, y_train)
```

```python
# Make predictions on the test set
y_pred = lgb_model.predict(x_test)
```

```python
# Evaluate the model
r2 = r2_score(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
```

```python
print(f"R^2: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red')
plt.xlabel('Actual LAI')
plt.ylabel('Predicted LAI')
plt.title(f'Best-Fit Model: LightGBM\nR-squared: {r2:.4f}, RMSE: {rmse:.4f}, MAE: {mae:.4f}')
plt.show()
```

```python
# Get feature importance scores
feature_importance = lgb_model.feature_importances_
```

```python
# Create a DataFrame to hold feature names and their importance scores
feature_importance_df = pd.DataFrame({'Feature': x.columns, 'Importance': feature_importance})
```

```python
# Sort features by importance
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
```

```python
# Plot feature importance
plt.figure(figsize=(10, 6))
plt.bar(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xticks(rotation=45, ha="right")
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importance using LightGBM')
plt.tight_layout()
plt.show()
```

```python
# Hyperparameter tuning loop for LightGBM
np.random.seed(42)
for i in range(5, 100, 5):
    print(f"Trying model with {i} estimators...")
    lgb_model = lgb.LGBMRegressor(n_estimators=i, learning_rate=0.1)
    lgb_model.fit(x_train, y_train)
    y_pred = lgb_model.predict(x_test)
    r2 = r2_score(y_test, y_pred)
    print(f"Model R^2 score on test set: {r2:.4f}")
    print("")
```
